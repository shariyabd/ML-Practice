```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set_theme()     
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
```

```python
print(tips)
```

```python
sns.set_style("whitegrid")   # options: white, dark, whitegrid, darkgrid, ticks
sns.set_context("notebook")  # paper, notebook, talk, poster
sns.set_palette("viridis")   # or sns.color_palette("pastel")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="sex", size="size", sizes=(20,200))
plt.title("Scatter: bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips, x="size", y="total_bill", estimator="mean", ci=95)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="time", kind="scatter")
```

```python
sns.countplot(data=tips, x="day", hue="sex")
```
