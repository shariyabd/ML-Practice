```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker="o", linestyle="--", color="blue")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(categories, values, color="skyblue")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.barh(categories, values, color="orange")
plt.show()
```

```python
sizes = [30, 20, 40, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90, colors=["red","green","blue","yellow"])
plt.title("Pie Chart")
plt.show()
```

```python
x = np.random.randint(10, 50, 30)
y = np.random.randint(20, 80, 30)

plt.scatter(x, y, color="purple", marker="x", s=100)  # s = size
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
data = np.random.randn(1000)  # normal distribution

plt.hist(data, bins=30, color="green", alpha=0.7)
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.randint(1, 100, 50)
y = np.random.randint(1, 100, 50)
z = np.random.randint(1, 100, 50)

ax.scatter(x, y, z, c="green", marker="o", s=60)
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
```
