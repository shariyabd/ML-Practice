```python
import matplotlib.pyplot as plt;
```

```python
import numpy as np;
```

```python
x = [1,2,3,4,5]
y = [2,4,6,7,9]

plt.plot(x,y, marker="o",color="blue", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot")
plt.show()
```

```python
categories = ["A","B","C","D"];
values = [5,10,20,25];

plt.bar(categories,values, color="blue")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar chart")
plt.show()
```

```python
plt.barh(categories,values,color="green")
plt.show()
```

```python
sizes = [30, 20, 40, 10]
labels = ["A", "B", "C", "D"]
```

```python
plt.pie(sizes, labels=labels,autopct="%1.1f%%", startangle=90, colors=["red","green","yellow","blue"])
plt.show()
```
